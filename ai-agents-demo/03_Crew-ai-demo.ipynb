```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI
from crewai import Agent, Task, Crew
#from langchain_community.llms import Ollama
from google.colab import userdata
# Ensure your API Key is set
os.environ["GOOGLE_API_KEY"] = userdata.get('GEMINI_API_KEY')

# ---------------------------
# LLM (Gemma 2B via Ollama)
# ---------------------------
#llm = Ollama(
#    model="gemma:2b",
#    temperature=0.3
#)
# 2. Initialize the Gemini Model
# Use gemini-2.0-flash or gemini-1.5-pro
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0
)

# ---------------------------
# Research Agent
# ---------------------------
research_agent = Agent(
    role="AI Research Analyst",
    goal="Research a given topic and summarize key insights clearly",
    backstory=(
        "You are an experienced research analyst who specializes in "
        "breaking down complex technical topics into simple explanations."
    ),
    llm=llm,
    verbose=True
)

# ---------------------------
# Research Task
# ---------------------------
research_task = Task(
    description=(
        "Research the topic: 'ReAct pattern in AI agents'.\n\n"
        "Cover:\n"
        "1. What it is\n"
        "2. Why it is used\n"
        "3. A simple example use case\n"
    ),
    expected_output=(
        "A concise research note with headings and bullet points, "
        "written in simple language."
    ),
    agent=research_agent
)

# ---------------------------
# Crew
# ---------------------------
crew = Crew(
    agents=[research_agent],
    tasks=[research_task],
    verbose=True
)

# ---------------------------
# Run
# ---------------------------
result = crew.kickoff()
print("\nFinal Output:\n")
print(result)
```
